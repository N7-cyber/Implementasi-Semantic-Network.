```python
graph = {
    # --- KELAS UMUM ---
    "mahasiswa": {
        "mengambil": "matkul"
    },
    "dosen": {
        "mengajar": "matkul"
    },

    # --- INSTANCE & ATRIBUT SPESIFIK ---
    "rafi": {
        "is-a": "mahasiswa",
        "mengambil": "kecerdasan-buatan",
        "diajar-oleh": "ali",
        "kuliah-di": "unusida",
        "jurusan": "informatika",
        "angkatan": "2024",
        "kelas": "if24b"
    },
    "ali": {
        "is-a": "dosen",
        "mengajar-mhs": "rafi",
        "mengajar-matkul": "kecerdasan-buatan",
        "gelar": "m.kom",
        "nidn": "no. nidn 12345",
        "prodi": "informatika"
    },
    "kecerdasan-buatan": {
        "is-a": "matkul",
        "jadwal": "jam matkul 17.50", # Bisa diisi jam aslinya
        "ruangan": "r404"
    }
}
```

```python
def get_property(node, prop):
    if node not in graph:
        return "objek tidak ditemukan"

    if prop in graph[node]:
        return graph[node][prop]

    if "is-a" in graph[node]:
        parent = graph[node]["is-a"]
        return get_property(parent, prop)

    return "tidak diketahui"
```

```python
def jawab_pertanyaan(kalimat):
    kata = kalimat.lower().split()

    try:
        objek = kata[0]

        if "mengambil" in kata:
            return get_property(objek, "mengambil")
        elif "adalah" in kata or "is-a" in kata:
            return get_property(objek, "is-a")
        elif "kuliah" in kata:
            return get_property(objek, "kuliah-di")
        elif "jadwal" in kata:
            return get_property(objek, "jadwal")
        elif "ruangan" in kata:
            return get_property(objek, "ruangan")
        elif "mengajar" in kata:
             return get_property(objek, "mengajar-matkul")
        else:
            properti = kata[-1]
            return get_property(objek, properti)

    except:
        return "error / input tidak valid"
```

```python
while True:
    tanya = input("\nTanya (ketik 'exit' untuk keluar): ")

    if tanya.lower() == "exit":
        break

    jawaban = jawab_pertanyaan(tanya)
    print("Jawaban:", jawaban)
```
